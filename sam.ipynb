```python
df['Attrition']=df['Attrition'].map({'Yes':1,'No':0})
df.head()
```

```python
y = df['Attrition']
attrition_count = y.value_counts()
attrition_count
```

```python
# Use seaborn.barplot to create the bar chart
import seaborn as sns
sns.barplot(x=attrition_count.index, y=attrition_count.values)
plt.xlabel('Attrition') # Add x-axis label
plt.ylabel('Count') # Add y-axis label
plt.title('Distribution of Attrition') # Add a title
plt.show() # Display the plot
```

```python
print("Columns before dropping:", df.columns.tolist())
df = df.drop(columns=['EmployeeCount','Over18','StandardHours'])
print(df.columns.tolist())
df.info()
```

```python
x=df.drop(['Attrition'],axis=1)
x=pd.get_dummies(x,drop_first=True).astype(int)
x.head()
```

```python
feature_name= x.columns.tolist()
print("Feature Names:\n", feature_name)
```

```python
import pickle
with open('feature_name.pkl', 'wb') as f:
    pickle.dump(feature_name,f)   
```

```python
#defining feature and target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#SMOTE + Tomek
from sklearn.utils.extmath import randomized_svd
from imblearn.combine import SMOTETomek   
smote_tomek = SMOTETomek(random_state=42)
# Resampling the training data
X_train_resampled, y_train_resampled = smote_tomek.fit_resample(X_train, y_train)
```

```python
# Scaling the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_resampled = scaler.fit_transform(X_train_resampled)
X_test = scaler.transform(X_test)
```

```python
import pickle

with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
# training the model
from sklearn.linear_model import LogisticRegression


model = LogisticRegression(class_weight='balanced',max_iter=10000)
model.fit(X_train_resampled, y_train_resampled)
```

```python
#To find predicted target
y_predict= model.predict(X_test)
```

```python
#applying sigmoid probability
predict_prob=model.predict_proba(X_test)
```

```python
y_predict_th = np.where(model.predict_proba(X_test)[:,1]>0.5,1,0) # Threshold is 0.5
pd.crosstab(y_test,y_predict_th)
```
